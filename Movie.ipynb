## The list is in string formate so lets convert it using 'Abstract Syntax Trees' (ast) library

```python
import ast
```

```python
def generes_extraction(obj):
    genere=[]
    for i in ast.literal_eval(obj):
        genere.append(i['name'])
    return genere
```

```python
df['genres']=df['genres'].apply(generes_extraction)
```

```python
df.head()
```

## Keywodrds

```python
df['keywords'][0]
```

```python
def extract_keywords(obj):
    keywords=[]
    for i in ast.literal_eval(obj):
        keywords.append(i['name'])
    return(keywords)
```

```python
df['keywords']=df['keywords'].apply(extract_keywords)
```

```python
df.head()
```

## Cast

```python
df['cast'][0]
```

```python
def extract_cast(obj):
    stop = 0
    casts = []
    for i in ast.literal_eval(obj):
        if stop > 3:
            break
        casts.append(i['name'])
        stop += 1
    return casts
```

```python
df['cast']=df['cast'].apply(extract_cast)
```

```python
df.head()
```

## Crew

```python
df['crew'][0]
```

```python
for i in ast.literal_eval(df['crew'][0]):
    if i['job']=='Director':
        print(i['name'])
```

```python
def extract_director(obj):
    dic_name=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            dic_name.append(i['name'])
    return dic_name
```

```python
df['crew']=df['crew'].apply(extract_director)
```

## 'Writer' and 'Writing' has same meaning so it doesn't make sense for keeping both. To remove letter lik this lets use  spaCy Lemmatizer

```python
new_df.head(2)
```

```python
import spacy

nlp = spacy.load("en_core_web_sm")

def lemmatize_text(text):
    doc = nlp(text)
    return " ".join([token.lemma_ for token in doc if token.is_alpha])
```

```python
new_df = new_df.copy()
new_df['tags'] = new_df['tags'].apply(lemmatize_text)
```

## Rechaekaing

```python
count_vectorizer=CountVectorizer(max_features=5000,stop_words='english')
vectors=count_vectorizer.fit_transform(new_df['tags']).toarray()
print(len(count_vectorizer.get_feature_names_out()))
for i in count_vectorizer.get_feature_names_out():
    print(i)
```

## Calculate the cosine distance for vactorization

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cs=cosine_similarity(vectors)
```

```python
cs[0]
```

```python
import pickle 

with open('cosinesimiliraty.pkl','wb') as f:
    pickle.dump(cs, f)
    
```

```python
def find_reccomendation(movie):
    movie_index=new_df[new_df['title']==movie].index[0]
    distace=cs[movie_index]
    shortest_distance_movies=sorted(list(enumerate(distace)),reverse=True, key= lambda x: x[1])[1:8]

    for i in shortest_distance_movies:
        
        print(new_df.iloc[i[0]]['title'])
    
```
